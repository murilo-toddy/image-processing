```python
def img_brightness(img_in, amount_bright=0):
  img_out= img_in.astype(float)+amount_bright
  img_out[img_out > 255] = 255
  img_out[img_out < 0] = 0
  img_out = img_out.astype(np.uint8)
  return img_out
```

```python
img4 = img_brightness(img, 60)
```

```python
fig = plt.figure(figsize=(6,6))
plt.subplot(221); plt.imshow(img, cmap="gray")
plt.subplot(222); plt.imshow(img2, cmap="gray")
plt.subplot(223); plt.imshow(img3, cmap="gray")
plt.subplot(224); plt.imshow(img4, cmap="gray")
```

```python
img4 = img_brightness(img, -100)
```

```python
fig = plt.figure(figsize=(6,6))
plt.subplot(221); plt.imshow(img, cmap="gray")
plt.subplot(222); plt.imshow(img2, cmap="gray")
plt.subplot(223); plt.imshow(img3, cmap="gray")
plt.subplot(224); plt.imshow(img4, cmap="gray")
```

```python
def img_brightness2(img_in, amount_bright=0):
  img_out = img_in.astype(float)+amount_bright

  # mapeia para 0-1
  img_out = (img_out - np.min(img_out[:])) / (np.max(img_out[:]) - np.min(img_out[:]))
  
  # mapeia para 0-255 e converte pra uint8
  img_out = (img_out*255).astype(np.uint8)
  return img_out
```

```python
img5 = img_brightness(img, -150)
img6 = img_brightness(img, 150)

fig = plt.figure(figsize=(6,6))
plt.subplot(221); plt.imshow(img, cmap="gray")
plt.subplot(223); plt.imshow(img5, cmap="gray"); plt.title('-150')
plt.subplot(224); plt.imshow(img6, cmap="gray"); plt.title('+150')
```
